```python
#import libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
#Search for target protein
target = new_client.target
target_query = target.search('butyrylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
## Retrieve bioactivity data of AChE in Homo Sapiens
selected_target = 'CHEMBL1914'

activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target)
df = pd.DataFrame.from_dict(res)
df
```

```python
#save the bioactivity data of AChE
df.to_csv('/content/drive/MyDrive/datasets/urop/bche/data/bioactivity_bche.csv', index = False)
```

```python
#removing records with missing IC50 values and canonical smiles
df_curated = df[df.standard_value.notna()]
df_curated = df_curated[df_curated.canonical_smiles.notna()]
df_curated = df_curated.drop_duplicates(['canonical_smiles'])
df_curated.shape
```

```python
# keep all the rows with type = IC50, pIC50, Log IC50, -Log IC50
df_curated = df_curated[df_curated['type'].isin(['IC50', 'pIC50', 'Log IC50', '-Log IC50'])]
df_curated
```

```python
# each type, print the possible units
for type in df_curated['type'].unique():
  print(type, df_curated[df_curated['type'] == type]['units'].unique())
```
